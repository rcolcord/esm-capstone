```python
import pandas as pd
import os
import re
import itertools
from src import util
```

```python
path = "data/renewables/"
profiles = {}

coord_pattern = re.compile(r"(-?\d+\.?\d*)[_,](-?\d+\.?\d*)")

# Step 1: load raw files into nested dict
for filename in os.listdir(path):
    if not filename.endswith(".csv"):
        continue

    filepath = os.path.join(path, filename)

    if "pv" in filename.lower():
        resource = "solar"
    elif "wind" in filename.lower():
        resource = "wind"
    else:
        continue

    match = coord_pattern.search(filename)
    if not match:
        print(f"Could not parse coordinates from {filename}")
        continue

    lat = float(match.group(1))
    lon = float(match.group(2))
    coord = (lat, lon)

    df = pd.read_csv(filepath, comment="#")

    # Clean time/index/column names
    df["time_utc"] = pd.to_datetime(df["time"])
    df = df.set_index("time_utc")
    df = df.rename(columns={"electricity": f"{resource}_cf"})

    # Drop unused columns only if they exist
    drop_cols = [c for c in ["local_time", "time"] if c in df.columns]
    df = df.drop(columns=drop_cols)

    # Keep only the CF column
    df = df[[f"{resource}_cf"]]

    if coord not in profiles:
        profiles[coord] = {}

    profiles[coord][resource] = df

# Step 2: combine wind + solar into one dataframe per coordinate
combined_profiles = {}

for coord, coord_dict in profiles.items():
    if "wind" not in coord_dict or "solar" not in coord_dict:
        print(f"Skipping {coord}: missing wind or solar")
        continue

    combined = coord_dict["wind"].join(coord_dict["solar"], how="inner")

    # Optional sanity checks
    if combined.isna().any().any():
        print(f"Warning: missing values after join at {coord}")

    combined_profiles[coord] = combined.sort_index()

profiles = combined_profiles
```

```python
# ---------- Find highest/lowest correlation trios ----------

all_trios = list(itertools.combinations(profiles.keys(), 3))

trio_scores = []
for trio in all_trios:
    score = util.trio_correlation_score(trio, profiles)
    trio_scores.append((trio, score))

# Sort low to high
trio_scores.sort(key=lambda x: x[1])

low_trio, low_score = trio_scores[0]
high_trio, high_score = trio_scores[-1]

print("Lowest-correlation trio:", low_trio, "score =", low_score)
print("Highest-correlation trio:", high_trio, "score =", high_score)
```

```python
low_trio_df = util.build_trio_timeseries(low_trio, profiles)
high_trio_df = util.build_trio_timeseries(high_trio, profiles)

# Save to CSV
low_trio_df.to_csv("data/low_correlation_trio_timeseries.csv")
high_trio_df.to_csv("data/high_correlation_trio_timeseries.csv")

print("Saved low_correlation_trio_timeseries.csv")
print("Saved high_correlation_trio_timeseries.csv")
```

```python
# Load CAISO load data

path = "data/load/"

load_raw = pd.read_excel(path + "historicalemshourlyloadfor2023.xlsx")

# Clean load data
# 8 hours of load data from 2022 has already been manually added to account for UTC time shift

load = load_raw

load["Date"] = pd.to_datetime(load["Date"])
load["HR"] = pd.to_numeric(load["HR"])

# CAISO-style hour ending timestamps:
# HR 1 -> 01:00
# HR 24 -> next day 00:00
load["local_time"] = load["Date"] + pd.to_timedelta(load["HR"], unit="h")

# Mark duplicated fall-back hour:
# first 01:00 = DST, second 01:00 = standard time
ambiguous = pd.Series(False, index=load.index)

dupes = load["local_time"].duplicated(keep=False)
for ts, idx in load.loc[dupes].groupby("local_time").groups.items():
    idx = list(idx)
    if len(idx) != 2:
        raise ValueError(f"Expected 2 rows for duplicated time {ts}, got {len(idx)}")
    ambiguous.loc[idx[0]] = True
    ambiguous.loc[idx[1]] = False

load["time_utc"] = load["local_time"]

# Localize to Pacific, then convert to UTC
load.index = (
    load["time_utc"]
    .dt.tz_localize("America/Los_Angeles", ambiguous=ambiguous.to_numpy())
    .dt.tz_convert("UTC")
)

# Convert to hour-beginning
load.index = load.index - pd.Timedelta(hours=1)

load["load_mw"] = load["CAISO"]

# cleanup
load = load.drop(columns=["Date", "HR", "local_time", "PGE", "SCE", "SDGE", "VEA", "CAISO", "time_utc"])

# limit to 2023 UTC data
load = load.loc[
    (load.index >= "2023-01-01 00:00:00+00:00") &
    (load.index <= "2023-12-31 23:00:00+00:00")
]


```

```python
# Save clean load data

load.to_csv("data/clean_load.csv")
```
