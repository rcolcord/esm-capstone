```python
import numpy as np
import pandas as pd
import pypsa

import sys
from pathlib import Path

sys.path.insert(0, str(Path("src")))
```

```python
from src import visualize_results
import matplotlib.pyplot as plt

# Load and inspect
results = visualize_results.load_core_results("results")
visualize_results.check_columns(results)

summary = visualize_results.make_summary_table(results)
display(summary)

# Paper figures
visualize_results.set_paper_style()

# Main paper panel
fig, axes = visualize_results.plot_paper_main_panel(results)
plt.show()

# Alternative paper panel
fig, axes = visualize_results.plot_paper_alt_panel(results)
plt.show()

# Individual useful paper plots
fig, ax = plt.subplots()
visualize_results.plot_system_cost(results, ax=ax)
plt.show()

fig, ax = plt.subplots()
visualize_results.plot_curtailment(results, ax=ax)
plt.show()

fig, ax = plt.subplots()
visualize_results.plot_transmission_build(results, ax=ax)
plt.show()

fig, ax = plt.subplots()
visualize_results.plot_ldes_power(results, ax=ax)
plt.show()

fig, ax = plt.subplots()
visualize_results.plot_ldes_energy(results, ax=ax)
plt.show()

fig, ax = plt.subplots()
visualize_results.plot_backup_capacity(results, ax=ax)
plt.show()

fig, ax = plt.subplots(figsize=(8.5, 4.8))
visualize_results.plot_storage_comparison(results, ax=ax)
plt.show()

fig, ax = plt.subplots(figsize=(7, 6))
visualize_results.plot_ldes_vs_transmission(results, ax=ax)
plt.show()

fig, ax = plt.subplots(figsize=(10, 5))
visualize_results.plot_generation_mix_bars(results, ax=ax)
plt.show()

# Optional cost figures
if {
    "generator_capex_total",
    "storage_capex_total",
    "link_capex_total",
    "generator_opex_total",
}.issubset(results.columns):
    fig, ax = plt.subplots(figsize=(10, 5))
    visualize_results.plot_cost_components(results, ax=ax)
    plt.show()

if {"total_capex", "total_opex"}.issubset(results.columns):
    fig, ax = plt.subplots(figsize=(8.5, 4.8))
    visualize_results.plot_capex_vs_opex(results, ax=ax)
    plt.show()

if "cost_reconstruction_error" in results.columns:
    fig, ax = plt.subplots(figsize=(8.5, 4.5))
    visualize_results.plot_cost_reconstruction_error(results, ax=ax)
    plt.show()

# Optional faceted comparison
g = visualize_results.plot_longform_facets(
    results,
    metrics=[
        "system_cost_per_mwh",
        "transmission_mw",
        "ldes_mw",
        "vres_curtailment_mwh",
    ],
    metric_labels={
        "system_cost_per_mwh": "System Cost ($/MWh)",
        "transmission_mw": "Transmission (MW)",
        "ldes_mw": "LDES (MW)",
        "vres_curtailment_mwh": "Curtailment (MWh)",
    },
)
plt.show()

# Save all paper figures
visualize_results.create_paper_figures(results, outdir="figures/paper")


# Powerpoint figures
visualize_results.set_slide_style()

fig, axes = visualize_results.plot_slide_main_panel(results)
plt.show()

fig, ax = plt.subplots()
visualize_results.plot_transmission_build(results, ax=ax)
plt.show()

fig, ax = plt.subplots()
visualize_results.plot_curtailment(results, ax=ax)
plt.show()

fig, ax = plt.subplots()
visualize_results.plot_ldes_energy(results, ax=ax)
plt.show()

fig, ax = plt.subplots(figsize=(8.5, 5))
visualize_results.plot_storage_comparison(results, ax=ax)
plt.show()

fig, ax = plt.subplots(figsize=(7.5, 6))
visualize_results.plot_ldes_vs_transmission(results, ax=ax)
plt.show()

fig, ax = plt.subplots()
visualize_results.plot_backup_capacity(results, ax=ax)
plt.show()

# Save all pp figures
visualize_results.create_slide_figures(results, outdir="figures/slides")
```
